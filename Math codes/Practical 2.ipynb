```python
from matplotlib.pyplot import *
x1=[1920,1930,1940,1950,1960,1970,1980,1990,2000,2010]
y1=[9.8,12,8,7.2,6.9,7,6.5,6.2,5.5,6.3]
plot(x1,y1,label="line 1")
xlabel('year')
ylabel('Unemployed rate')
title('LINE GRAPH')
legend()
show()
```

```python
from matplotlib.pyplot import*
x1=[3,1,1,6]
y1=[1,6,3,1]
plot(x1,y1,label="line 1")
x2=[1,8,1,3]
y2=[1,3,1,8]
plot(x2,y2,label="line 2")
x3=[0,1,0,8]
y3=[0,8,0,1]
plot(x3,y3,label="line 3")
xlabel('x-axis')
ylabel('y-axis')
title('LINE GRAPH')
legend()
show()
```

```python
from matplotlib.pyplot import*
x=[130,120,135,130,150,80]
y=[1,2,3,4,5,6]
tick_label=['mon','thu','wed','thur','fri','sat']
bar(y,x,tick_label=tick_label,width=0.5,color=['yellow','red'])
xlabel('days')
ylabel('absent')
title('BAR GRAPH')
legend()
show()
```

```python
from matplotlib.pyplot import*
x=[175,325,100,150,250,525]
range=(100,600)
intervals=6
hist(x,intervals,range,width=80,histtype='bar')
xlabel('no.of illterature')
ylabel('age group')
title('HISTROGRAM')
legend()
show()
```
